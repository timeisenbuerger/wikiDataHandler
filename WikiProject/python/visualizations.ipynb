```python
# read the csv
import pandas as pd

data_filepath = '../wikiData.csv'
data = pd.read_csv(data_filepath, delimiter = ';', header = None)

data.head()
```

```python
data.columns = ['features', 'language']
data.head()
```

```python
type_systems = ['static', 'dynamic', 'hybrid', 'strong', 'weak', 'inferred', 'manifest', 'partially inferred', 
                'nominal', 'structural', 'nominative', 'parametric', 'optional', 'duck', 'safe', 'dependent', 
                'polymorphic'] # except 'none'

# add new columns
def add_columns(data):    
    for index, row in data.iterrows():
        for type_system in type_systems: 
            if type_system in row['features']:
                data.loc[index, type_system] = 1
            else:
                data.loc[index, type_system] = 0

add_columns(data)
data.head()
```

```python
# drop features column
data = data.drop('features', axis=1)

# remove "(programming language)"
data['language'] = data['language'].str.replace(' \(programming language\)', '')
data['language'] = data['language'].str.replace(' \(parallel scripting language\)', '')

data.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

# type system distribution
counts = data.sum(axis = 0).drop('language').tolist()
plt.figure(figsize=(25,10))
colors = ['navajowhite', 'navajowhite', 'navajowhite', 
          'skyblue', 'skyblue', 
          'lightgreen', 'lightgreen', 'lightgreen', 
          'violet', 'violet', 
          'lightgrey', 'lightgrey', 'lightgrey', 'lightgrey', 'lightgrey', 'lightgrey', 'lightgrey']
plt.bar(type_systems, counts, color = colors)
```

```python
# most popular languages according to TIOBE index
pop_lang = ['Java', 'C', 'Python', 'C++', 'Visual Basic .NET', 'JavaScript', 'C Sharp', 'PHP', 'SQL',
            'Objective-C', 'R', 'Swift', 'Go', 'Ruby']

pop_data = data[data['language'].isin(pop_lang)]
pop_data.head(len(pop_lang))
```

```python
pop_counts = pop_data.sum(axis = 0).drop('language').tolist()
plt.figure(figsize=(25,10))
plt.bar(type_systems, pop_counts, color = colors)
```

```python
import seaborn as sns
import numpy as np

# correlations
corrmat = data.corr(method = 'pearson')

# Generate a mask for the upper triangle
mask = np.zeros_like(corrmat, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

#Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

sns.heatmap(corrmat, mask = mask, cmap = 'PRGn')
```

```python
# most common type systems: strong, dynamic, static, hybrid, inferred
total_lang = data.shape[0]
print("There are a total of " + str(total_lang) + " programming languages. ")

# strong & safe
total_strong = data['strong'].sum()
strong_safe = data.loc[data['strong'] == 1, 'safe'].sum()
percent = '%.2f'%(strong_safe  * 100/ total_strong)
print(str(strong_safe) + " of " + str(total_strong) + " languages with a strong type system are also safe. This is " 
      + str(percent) + " %.")


# strong & static
total_strong = data['strong'].sum()
strong_static = data.loc[data['strong'] == 1, 'static'].sum()
percent = '%.2f'%(strong_static  * 100/ total_strong)
print(str(strong_static) + " of " + str(total_strong) + " languages with a strong type system are also static. This is " 
      + str(percent) + " %.")

# strong & dynamic
total_strong = data['strong'].sum()
strong_dynamic = data.loc[data['strong'] == 1, 'dynamic'].sum()
percent = '%.2f'%(strong_dynamic  * 100/ total_strong)
print(str(strong_dynamic) + " of " + str(total_strong) + " languages with a strong type system are also dynamic. This is " 
      + str(percent) + " %.")

# dynamic & strong
total_dynamic = data['dynamic'].sum()
dynamic_strong = data.loc[data['dynamic'] == 1, 'strong'].sum()
percent = '%.2f'%(dynamic_strong  * 100/ total_dynamic)
print(str(dynamic_strong) + " of " + str(total_dynamic) + " languages with a dynamic type system are also strong. This is " 
      + str(percent) + " %.")

# static & inferred
total_static = data['static'].sum()
static_inferred = data.loc[data['static'] == 1, 'inferred'].sum()
percent = '%.2f'%(static_inferred  * 100/ total_static)
print(str(static_inferred) + " of " + str(total_static) + " languages with a static type system are also inferred. This is " 
      + str(percent) + " %.")

# inferred & structural
total_inferred = data['inferred'].sum()
inferred_structural = data.loc[data['inferred'] == 1, 'structural'].sum()
percent = '%.2f'%(inferred_structural  * 100/ total_inferred)
print(str(inferred_structural) + " of " + str(total_inferred) + " languages with a inferred type system are also structural. This is " 
      + str(percent) + " %.")
```
